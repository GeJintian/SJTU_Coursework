```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('C:/Users/69009/Desktop/last/441/mid/demands.csv')
dataset
```

```python
df_d = pd.DataFrame(dataset, columns=["demands"])
df_t = pd.DataFrame(dataset, columns=["time"])
df_d
```

```python
plt.scatter(df_t,df_d)
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(df_t, df_d)
model.fit(df_t, df_d)
results = model.predict(df_t)
```

```python
plt.scatter(df_t,df_d)
plt.plot(df_t,results)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import ensemble
from sklearn import tree

params={'n_estimators': 1, 'max_depth': 1, 'learning_rate': 1, 'loss': 'ls'}
model = ensemble.GradientBoostingRegressor(**params)
model.fit(df_t,df_d)
results = model.predict(df_t)
plt.scatter(df_t,df_d)
plt.plot(df_t,results)
plt.show()
```
